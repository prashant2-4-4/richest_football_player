```python
from selenium import webdriver
import os
import time
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
```

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
url= 'https://ceoworld.biz/2021/01/27/worlds-richest-soccer-players-2021/'
driver.get(url)
driver.implicitly_wait(5)
all_name = driver.find_elements(By.XPATH , '//*[@id="tablepress-536"]/tbody/tr/td[2]')
```

```python
player_name = []
for t in all_name:
    print(t.text)
    player_name.append(t.text)
print(len(player_name))
```

```python
net_worth = driver.find_elements(By.XPATH ,'//*[@id="tablepress-536"]/tbody/tr/td[3]' )
net_worth_of_player = []
for t in net_worth:
    print(t.text)
    net_worth_of_player.append(t.text)
print(len(net_worth_of_player))
```

```python
profession = driver.find_elements(By.XPATH , '//*[@id="tablepress-536"]/tbody/tr/td[5]')
profession_player = []
for t in profession:
    print(t.text)
    profession_player.append(t.text)
print(len(profession_player))
```

```python
nationality = driver.find_elements(By.XPATH , '//*[@id="tablepress-536"]/tbody/tr/td[4]')
nationality_player = []
for t in nationality:
    print(t.text)
    if(t.text == '-'):
        nationality_player.append('Egypt')
    else:
        nationality_player.append(t.text)
print(len(nationality_player))
```

```python
import pandas as pd
dict_ = {'player_name' : player_name , 'net_worth_by_2021' : net_worth_of_player , 'nationality' : nationality_player , 'profession_other_than_soccer' : profession_player}
richest_footballer = pd.DataFrame(dict_)
richest_footballer
```
